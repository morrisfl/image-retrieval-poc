```python
from annoy import AnnoyIndex
from torchvision import transforms

from src.data import Paris6kDataset
from src.models import CLIPModelPretrained
from src.utils import get_indexes, visualize_retrieval_results
```

```python
# Model
model = CLIPModelPretrained(model_name="ViT-B/32", pretrained_model="laion2b_s34b_b79k")
transform = model.preprocess

# Dataset
dataset = Paris6kDataset(root="../../dataset/paris6k", transform=transform)
query_dict = dataset.query_data
gallery_images = dataset.gallery_data

# Stored embeddings
embeddings = AnnoyIndex(512, "angular")
embeddings.load("../../embeddings/index25_clip_ViT-B-32_laion2B_paris6k.ann")
```

```python
query_label = "eiffel"
query_img_path = query_dict[query_label][0]

indexes = get_indexes(model=model, img_path=query_img_path, embeddings=embeddings, top_k=6, transform=transform)

visualize_retrieval_results(query_label=query_label, img_path=query_img_path, gallery_img=gallery_images, indexes=indexes)
```
